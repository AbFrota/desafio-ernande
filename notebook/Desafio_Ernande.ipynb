```python
import pandas as pd
```

```python
df = pd.read_csv('LOG_Ambiente.csv',sep =";")
```

```python
df.head()
```

```python
df['cenario'] = 0
df.head()
```

```python
df['cenario'][df['temperature_in'] > 27] = 1
df['cenario'][df['temperature'] < 20] = 1
df['cenario'][df['temperature'] > 24] = 1
df['cenario'][df['humidity'] < 45] = 1
df['cenario'][df['humidity'] > 55] = 1
df['cenario'][df['gas'] > 500] = 1
```

```python
df[df['cenario'] == 0]
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace = True)

```

```python
df.to_csv('LOG_Ambiente.csv',sep =";", index = False)
```

```python
df.drop('Unnamed: 0', inplace =  True, axis = 1)
```
